#### Обратную матрицу вычислить нельзя потому что det(B) равен нулю

```python
np.random.seed(42)
a =np.random.random(16)
C = a.reshape((4,4))
C = C * 10
```

```python
D = B + C
np.linalg.det(D)
np.linalg.matrix_rank(D)
D_inv = np.linalg.inv(D)
D_inv
```

```python
for i in range(len(D_inv)):
    for j in range(len(D_inv[i])):
        if D_inv[i][j] > 0:
            D_inv[i][j] = 1
        else:
            D_inv[i][j] = 0
```

```python
D_inv
```

```python
E = np.where(D_inv,(B,C))
```
